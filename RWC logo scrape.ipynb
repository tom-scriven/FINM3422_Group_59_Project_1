```python
import requests
from io import BytesIO
from PIL import Image

# 1) Set the image URL
img_url = (
    "https://upload.wikimedia.org/wikipedia/commons/thumb/"
    "c/cc/Reliance_Worldwide_Corporation_logo.svg/"
    "512px-Reliance_Worldwide_Corporation_logo.svg.png?20231224033219"
)

# 2) Download the image
resp = requests.get(img_url)
resp.raise_for_status()
img_bytes = resp.content

# Save the image for the pdf report
with open("rwc_logo.png", "wb") as f:
    f.write(img_bytes)
image = Image.open(BytesIO(img_bytes))
display(image)
```
