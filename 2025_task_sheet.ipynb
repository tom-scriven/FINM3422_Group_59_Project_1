# FINM3422 Equity Research Assignment
### Task:
You are an equity research analyst and you have been tasked with developing and implementing a program that produces an equity research report as a pdf file for an assigned company in the ASX200. As a group, you will be randomly assigned a ticker to report on.

One team member will be the leader, who will create a new **private** GitHub repo and invite team members as collaborators, with each team member creating a different enhancement to the program which will become included in the research report pdf. Do not set your repo to public visibility at any time, before or after submission.

In your blank repo create a new notebook (.ipynb) file named _analyst\_report.ipynb_. You will develop your main program inside this notebook file.
### Features and Marking
The program should produce an attractive and well-formatted report. The features that your program can have, along with the marks assigned for each feature are:
- Create a pdf file of the final research report **[20%]**
- Key ratios and financial metrics informed through the use of an API, and computed in functions that have been created, saved into, and imported from a .py file **[20%]**
- Include commentary imported from a .txt file **[10%]**
- An extract from a DCF model built in Excel **[10%]**
- A chart showing price history for the stock, historical price targets of the research (taken from a CSV file), and the history of the ASX200 **[20%]**
- A relevant image or table scraped from a website for the company **[10%]**

Marks will be awarded based on the quality of implementation of each of these features. The content of your research needs to be relevant (i.e., you need to produce a relevant commentary of the stock and an accurate DCF evaluation with a realistic target price), but your grade doesn't depend on the depth of your research. The focus of the assignment is on your ability to use Python and GitHub collaboratively, not on your financial analysis.

The final **10%** of marks will be awarded for the overall quality of your code. You will lose points here for redundant, hard to follow, or poorly commented code.
### Individual Marks
Individual marks will be modified based on the total quality of each team member's contribution as recorded by their commits to the forked repo.
